```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(1000)
y = np.random.normal(size=1000)
z = np.random.normal(size=1000)
```

```python
plt.figure(figsize=(16,10))
### 1st subplot, top left corner
plt.subplot2grid((2,4),(0,0))
plt.plot(x,y,'+',alpha=0.5)
plt.title('boo')
plt.tick_params(axis='both', which='major', labelsize=14)

### 2nd subplot, top row, 2nd column
plt.subplot2grid((2,4),(0,1))
plt.hist(y,alpha=0.5)
plt.title('asdfa')
plt.tick_params(axis='both', which='major', labelsize=14)

### 3rd subplot, top row, 3rd column
plt.subplot2grid((2,4),(0,2))
plt.hist(z,alpha=0.5)
plt.title('sdgfs')
plt.tick_params(axis='both', which='major', labelsize=14)

### 4th subplot, top right corner
plt.subplot2grid((2,4),(0,3))
plt.plot(z,y,alpha=0.5)
plt.title('hdgn')
plt.tick_params(axis='both', which='major', labelsize=14)


### 5th subplot, bottom row
plt.subplot2grid((2,4),(1,0), colspan=4)
plt.plot(x,y,'+',alpha=0.5)
plt.tick_params(axis='both', which='major', labelsize=14)


### Overplotting an inset plot:
ax2 = plt.axes([.44, .82, .05, .05])
# The numbers in the array respectively represent: 
# initial point in X 
# initial point in Y
# width of the subplot in X
# width of the subplot in Y 
# All numbers in fraction of the total length of axes
ax2.plot(z,y,'.r')
```

```python
import numpy as np
from astropy.time import Time
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
x = 56124.0 + np.sort(np.random.rand(10))
y = np.random.rand(10)
```

```python
plt.plot(x,y,'.:')
```

```python
plt.figure()
plt.plot(x,y,'.:')

plt.xlim(56124 + (4.0/24.0), 56125)
xlim_low, xlim_high = plt.xlim()
xticks_pos = np.arange(xlim_low, xlim_high,(xlim_high-xlim_low)/4.0000000001)
# The following command will make labels, but will have date and time as lables
xticks_label = Time(Time(xticks_pos,format='mjd',scale='utc').iso,out_subfmt='date_hms').value
# If you just want hours, mins and secs, you can slice the labels above:
for i, label in enumerate(xticks_label):
    xticks_label[i] = label[-13:-4]

plt.xticks(xticks_pos,xticks_label,rotation=45)
plt.tight_layout()
```

```python
import numpy as np
import pymc3 as pm
import corner
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,50) + np.random.normal(scale=1,size=50)
y = np.linspace(-10,10,50) + np.random.normal(scale=0.5,size=50)
y_er = abs(np.random.normal(loc=1.0,scale=0.2,size=50))

plt.figure(figsize=(8,6))
plt.errorbar(x,y,y_er,fmt='.')
```

```python
def line_model(x,m,b):
    return m*x+b

def quad_model(x,a,m,b):
    return a*x**2+m*x+b
```

```python
with pm.Model() as linear_model:
    slope = pm.Uniform('slope',-20,20)
    intercept = pm.Uniform('intercept',-20,20)
    Y = pm.Normal('Y',mu=line_model(x,slope,intercept), sd=y_er, observed=y)
    trace1 = pm.sample(5000,tune=1000,chains=20,cores=6)

pm.traceplot(trace1)
pm.posteriorplot.plot_posterior(trace1)
corner_data = list(zip(*np.array([trace1.get_values('slope').transpose(),
                                  trace1.get_values('intercept').transpose()])))
cornplot = corner.corner(corner_data,quantiles=[0.16,0.5,0.84],labels = ['slope','intercept'])

plt.figure(figsize=(8,6))
plt_x = np.arange(-15,15,0.1)
plt.errorbar(x,y,y_er,fmt='.b',zorder=11,elinewidth=0.5)
plt.plot(plt_x, line_model(plt_x, 
                           np.median(trace1.get_values('slope')), 
                           np.median(trace1.get_values('intercept'))),'-k',zorder=10)

for i in np.random.randint(0,1000,size=200):
    plt.plot(plt_x, line_model(plt_x, 
                               trace1.get_values('slope')[i],
                               trace1.get_values('intercept')[i]),'r', lw=0.1, alpha=0.5)


pm.summary(trace1)
```

```python
with pm.Model() as quadratic_model:
    acc = pm.Uniform('acc',-20,20)
    slope = pm.Uniform('slope',-20,20)
    intercept = pm.Uniform('intercept',-20,20)
    Y = pm.Normal('Y',mu=quad_model(x,acc,slope,intercept), sd=y_er, observed=y)
    trace2 = pm.sample(5000,tune=1000,chains=20,cores=6)

pm.traceplot(trace2)
pm.posteriorplot.plot_posterior(trace2)
corner_data = list(zip(*np.array([trace2.get_values('acc').transpose(),
                                  trace2.get_values('slope').transpose(),
                                  trace2.get_values('intercept').transpose()])))
cornplot = corner.corner(corner_data,quantiles=[0.16,0.5,0.84],labels = ['acc','slope','intercept'])

plt.figure(figsize=(8,6))
plt_x = np.arange(-15,15,0.1)
plt.errorbar(x,y,y_er,fmt='.b',zorder=11,elinewidth=0.5)
plt.plot(plt_x, quad_model(plt_x, 
                           np.median(trace2.get_values('acc')),
                           np.median(trace2.get_values('slope')),
                           np.median(trace2.get_values('intercept'))),'-k',zorder=10)

for i in np.random.randint(0,1000,size=200):
    plt.plot(plt_x, quad_model(plt_x, 
                               trace2.get_values('acc')[i],
                               trace2.get_values('slope')[i],
                               trace2.get_values('intercept')[i]),'r', lw=0.1, alpha=0.5)


pm.summary(trace2)
```

```python
linear_model.name = 'linear'
quadratic_model.name = 'quad'

pm.compare({linear_model: trace1, quadratic_model: trace2})
```

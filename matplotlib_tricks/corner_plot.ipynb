# Making a Corner plot

Making a corner plot using the [Corner](https://corner.readthedocs.io/en/latest/) package. This is a useful visualization method when dealing with probability distributions that may be correlated (e.g., MCMC parameter posterior distributions).

Note that the Corner package can handle [Emcee hammer](http://dfm.io/emcee/current/) output out of the box (as it was originally desiged for outputs from that package by the same developer).

```python
import numpy as np
import matplotlib.pylab as plt
import corner
```

```python
%matplotlib notebook
```

```python
def data_for_corner(N=50000):
    a = np.random.normal(size=N)
    b = np.random.chisquare(10,size=N)
    c = np.random.normal(1.2,size=N)
    corner_data = np.array([a,b,c]).T
    return corner_data
```

```python
corner_data = data_for_corner()
cornerplot = corner.corner(corner_data,labels=['a','b','c'],quantiles=[0.16, 0.5, 0.84],show_titles=True)
cornerplot.savefig('corner.pdf')
```
